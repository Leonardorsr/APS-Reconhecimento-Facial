```python
import cv2
```

```python

#Preração da câmera
camera = cv2.VideoCapture(0)
detectorFace = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")

#Preparação do Eigenfaces
reconhedorEigen = cv2.face.EigenFaceRecognizer_create()
reconhedorEigen.read("Classificadores_de_Treino\classificardorEigen.yml")

#Preparação do Fisherfaces
reconhecerdorFisher = cv2.face.FisherFaceRecognizer_create()
reconhecerdorFisher.read("Classificadores_de_Treino\classificardorFisher.yml")
#Preparação do LBPH
reconhecerdorLBPH = cv2.face.LBPHFaceRecognizer_create()
reconhecerdorLBPH.read("Classificadores_de_Treino\classificadorLBPH.yml")


largura, altura = 220, 220
font = cv2.FONT_HERSHEY_COMPLEX_SMALL
IDCorreto = False  
```

```python
#Rodar esse código caso queira calcular a precisão:

IDCorreto = int(input("ID de quem será reconhecido: "))
```

```python
contador = 0
acertos = 0
precisao =0

while True:
  idlist = []
  confiancalist = []
  id1=0
  id2=0
  id3=0
  id4=0
  
  
  
  conectado, imagem = camera.read()
  imagemCinza = cv2.cvtColor(imagem,cv2.COLOR_BGR2GRAY)
  facesDetectadas = detectorFace.detectMultiScale(imagemCinza, scaleFactor=1.5,minSize=(100,100))
  
  for (x,y,l,a) in facesDetectadas:
    
    imagemFace = cv2.resize(imagemCinza[y:y + a, x:x + l], (largura, altura))
    
    cv2.rectangle(imagem, (x,y), (x + l, y + a), (0,0,255) , 2)
    
    #Reconhecimento Eigen
    idEigen, confiancaEigen = reconhedorEigen.predict(imagemFace)
    idlist.append(idEigen)
    confiancalist.append(confiancaEigen)
    
    #Reconhecimento Fisher
    idFisher,confiancaFisher = reconhecerdorFisher.predict(imagemFace)
    idlist.append(idFisher)
    confiancalist.append(confiancaFisher)
    
    #Reconhecimento LBPH
    idLBPH, confiancaLBPH = reconhecerdorLBPH.predict(imagemFace)
    idlist.append(idLBPH)
    confiancalist.append(confiancaLBPH)
    
    
    
    for id in idlist:
      if id == 1:
        id1+=1
      elif id ==2:
        id2+=1
      elif id ==3:
        id3=+1
      elif id == 4:
        id4+=4
    
    if id1>id2 and id1>id3 and id1>id4:
      nome = "Leonardo"
      id=1
    elif id2>id3 and id2 >id4:
      nome = "Felipe"
      id=2
    elif id3 > id4:
      nome = "Vinicius"
      id=3
    elif id4 > id1 and id4 > id2 and id4 > id3:
      nome = "Rodrigues"
      id=4

     
    confianca = sum(confiancalist) /3
  
  
    cv2.putText(imagem, str(id)+nome+str(confianca), (x,y+(a+30)), font, 2, (0,255,0))
    #Cálculo da precisão
    if IDCorreto == id:
      acertos = acertos + 1
    contador = contador + 1
    precisao = acertos/contador *100
      
    if IDCorreto != False:
      cv2.putText(imagem, str("")+str(precisao)+str("%"), (x,y+(a+60)), font, 2, (0,0,255))
      
    else:
      cv2.putText(imagem, str(confianca), (x,y+(a+30)), font, 2, (0,255,0))
  
      
      
  cv2.imshow("Face", imagem)
  cv2.waitKey(1)

  if cv2.waitKey(1) == ord('q'):
    break
camera.release()
cv2.destroyAllWindows()
```
